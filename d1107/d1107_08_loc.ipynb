데이터 선택  - loc (row값을 가져옴)

```python
# index이름으로 검색해서 그것에 해당되는 컬럼 가져오기
# index 1번 국어점수 출력
df.loc['1번','국어']
df.loc[['1번','2번'],'국어']  # 1번,2번 학생의 국어점수 출력
df.loc[['1번','2번','5번'],['영어','수학']]
df.loc['2번':'6번','국어':'사회'] #슬라이싱 
```
