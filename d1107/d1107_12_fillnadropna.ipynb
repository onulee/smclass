```python
df = pd.read_excel('score.xlsx',index_col='지원번호')
df
```

```python
df['학교'] = np.nan
```

```python
df
```

```python
# dropna : NaN데이터 삭제, 
# axis : 축 - index: row삭제,columns:column 삭제
# how default : any - NaN값이 1개라도 있으면 삭제 , all - 전부NaN일때 삭제됨.
df.dropna(axis='columns',how='all')
```

* fillna : NaN 데이터 해당값으로 채워줌
* dropna : NaN 데이터를 삭제
- axis : 축 index-row방향으로 삭제,columns-columns방향으로 삭제
- how : 어떤경우에 삭제할지 결정 any-1개라 
NaN값이 있을때 삭제,all-모두 NaN일때 삭제 

* np.nan : numpy라이브러리 NaN 입력
