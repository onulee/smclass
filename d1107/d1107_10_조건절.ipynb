조건

```python
import pandas as pd
df = pd.read_excel('score.xlsx',index_col='지원번호')
df
```

```python
# 조건
df['키']>=185
```

```python
# 조건식을 변수에 저장
filt = df['키']>=185
df[filt]    # 조건에 해당되는 값만 출력
df[~filt]   # 조건에 해당되지 않는 값만 출력
df[df['키']>=185]
df[~(df['키']>=185)]
```

```python
# 같은 결과값.
df[df['키']>=185]       # 키가 185이상 모든 컬럼 출력
df.loc[df['키']>=185]   # 키가 185이상 모든 컬럼 출력, 컬럼을 지정할수 있음.
df.loc[df['키']>=185,'이름']  # 키가 185이상, 이름컬럼만 출력
df['키']>=185   # True, False 의 값으로 출력
df.loc[df['키']>=185,['이름','키','SW특기']]
```

조건식

```python
# 조건식 and &, or | 
filt = (df['키']>=185) & (df['학교'] == '구로고')
df[filt]
```

```python
### or 조건식
df.loc[(df['키']<170) | (df['키']>200)]
```

```python
### 문자열(str) 조건  -> 김으로 시작되는 :startwith
### 문자엻마수 endwith,strip,replace
filt = df['이름'].str.startswith('강')
df[filt]
df[df['이름'].str.startswith('김')]

### 수 가 포함되어 있는 이름을 검색 : contains()
filt = df['이름'].str.contains('김')
df[~filt]

# lower : 소문자 변환
df['SW특기'].str.lower()
# upper : 대문자 변환
df['SW특기'].str.upper()

df['SW특기']
langs = ['Python','Java']
## isin() : 리스트에 있는 단어가 포함되어 있는지 체크
filt = df['SW특기'].isin(langs) 
df[filt]

### 소문자 취환 : isin() 확인
langs = ['python','java']
filt = df['SW특기'].str.lower().isin(langs)
df[filt]

```

```python
df['SW특기'].str.lower().isin(langs)
```

```python
## Java라는 글자 있는지 체크
## na = True : NaN True인식, na=False : NaN False 인식
df['SW특기'].str.contains('Java')
filt = df['SW특기'].str.contains('Java',na=True)
df[filt]


df.info()
```
