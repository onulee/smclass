평균 80점이상, 국어 90점이상을 출력하시오.


```python
import pandas as pd
df = pd.read_csv('students.csv',index_col='no')
```

```python
# 평균 80점이상, 국어 90점이상
filt = (df['avg']>=80) & (df['kor'] >= 90)
df[filt]
```

```python
### 평균 60점이상 또는 국어 70점이상
filt = (df['avg']>=60) | (df['kor']>=70)
df[filt]
```

```python
### 이름 t가 들어가면서, 평균 80점이상 출력하시오.
filt = (df['name'].str.contains('t')) & (df['avg']>=80)
df['name']
df[filt]['kor'].count()
df['name']
```
