### 데이터 슬라이싱

```python
# row의 데이터 가져오기 , 슬라이싱
df[0:3] # index 0부터 3까지 가져오기
df[:4]  # 처음부터 4까지 가져오기
df      # 전체 가져오기
df[3:]  # 3부터 끝까지 가져오기
df[['이름','학교','SW특기']][:3]  # 이름,학교,SW특기 컬럼의 위3개를 가져오기
```

```python
### row의 값을 가져오는 방법 2가지 , df 슬라이싱, df.iloc 슬라이싱
df[1:3]
df.iloc[1:3]
```
