```python
import matplotlib.pyplot as plt
# 한글설정
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Win용
# matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac용
matplotlib.rcParams['font.size'] = 15 # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
## x,y축의 데이터
x = [1,2,3]  # 리스트,튜플,DataFrame데이터 - 리스트형태,Series
y = [2,4,8]
yy = [1,8,3]
## 선그래프 생성
# linewidth : 선두께,color : 선색상
# marker : 지점표시 - o,x,v, markersize : 마커크기,markeredgecolor: 마커테두리선
# markerfacecolor : 마커색상
# alpha : 투명도설정, linestyle:라인스타일 (:점선, -- dash선, -.)
plt.plot(x,y, label='인원변동',color='pink', linewidth=5, marker='o',markersize=10,markeredgecolor='red',markerfacecolor='yellow')
# plt.plot(x,yy, label='비용변동',marker='v',markersize=10,markerfacecolor='yellow',linestyle='--',alpha=0.2)
# plt.plot(x,yy, label='비용변동',marker='v',ms=20,mec='blue',mfc='green',ls='--',alpha=0.4)

###### 설정별 단축표시 'bv:'
# b:선색상,v:마커표시,: 선모양 linestyle='none'
plt.plot(x,yy,'bv:',label='비용변동')
## 상단타이틀
plt.title("라인그래프 Line",size=30)  # size : 글자크기 변경

## 범례 : 그래프 선 설명문구, loc:범례위치 upper,lower,left,center,right
# plt.legend(loc='upper center')
plt.legend(loc=(0.3,0.1))  # 좌표값으로 위치 설정 가능
```
