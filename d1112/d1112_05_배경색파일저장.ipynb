```python
import matplotlib.pyplot as plt
# 한글설정
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Win용
# matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac용
matplotlib.rcParams['font.size'] = 15 # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
import pandas as pd
import random
df = pd.read_excel('score.xlsx',index_col='지원번호')
```

```python
# X축 : 지원번호, Y축 국어
# 선색상 : red, marker색상:yellow, 테두리:blue, 선타입 --

x = df.index
k = df['국어']
e = df['영어']
# 그래프 크기 조정, dpi : 그래프해상도 설정 해상도가 높을수록 이미지 크기 커짐.
# facecolor : 배경색설정
plt.figure(figsize=(20,5),facecolor='yellow')  # x축으로 5, y축으로는 20 크기조절
# plt.figure(figsize=(20,5),dpi=300)  # x축으로 5, y축으로는 20 크기조절
# b:선색상,v:마커표시,: 선모양
plt.plot(x,k,label='국어',marker='o',color='red',mfc='yellow',mec='blue',ms=15,ls='--')
plt.plot(x,e,label='영어',marker='v',ms=15)
plt.legend()

# 그래프 파일로 저장
# plt.savefig('graph2.png',dpi=300)
```
