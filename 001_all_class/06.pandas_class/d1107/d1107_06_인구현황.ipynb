```python
### 남자부분만 가져옴.
### 여자부분만 가져와서 출력하시오.
## 위3개를 제외하고, 행정기관 index, 여자부분만 가져옴.
## [0] ,천단위 제거, 타입을 int변경하시오.
## B,AB,AV
df_w = pd.read_excel('연령별인구현황01.xlsx',skiprows=3,index_col='행정기관',usecols='B,AB:AV')
df.columns
df_w.columns
df['0~4세']
df_w['0~4세.1']
# 여자컬럼과 남자컬럼명 다르기에 통일
df_w.columns = df.columns
df_w.columns

# 천단위,int타입변경
df_w.iloc[0] = df_w.iloc[0].str.replace(",","").astype(int)
df_w.iloc[0]
```
