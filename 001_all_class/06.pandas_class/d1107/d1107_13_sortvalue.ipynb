```python
import pandas as pd
import numpy as np
df = pd.read_excel('score.xlsx',index_col='지원번호')
df
```

```python
# index정렬, ascending=False : 역순정렬
df.sort_index(ascending=False)
```

```python
# sort_value : 컬럼의 값으로 정렬, ascending=False :역순정렬
df.sort_values('키',ascending=False)
df.sort_values('이름')
df.sort_values('국어')
```

```python
df.sort_values(['국어','영어'],ascending=False)
```

```python
df.sort_values(['국어','영어'],ascending=[False,True])
```

* sort_index : index값을 가지고 정렬
* sort_values : columns값을 가지고 정렬
