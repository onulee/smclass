```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# 한글설정
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Win용
# matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac용
matplotlib.rcParams['font.size'] = 15 # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False

df = pd.read_excel('score.xlsx',index_col='지원번호')
```

```python
df['학년'] = [3,3,2,1,1,3,2,2]
```

```python
import random
```

```python
## 산점도그래프
# sizes = np.random.rand(8)*1000
sizes = df['학년']*500
plt.title("산점도 그래프")
plt.xlabel('영어')
plt.ylabel('수학')
## s : 산점도 크기 변경, c = 산점도 표시 컬럼내용 색상 
## cmap 색상표사용,
plt.scatter(df['영어'],df['수학'],s=sizes,c=df['학년'],cmap='plasma', alpha=0.5)
## colorbar : 색상바 생성 - 산점도색상의 학년표시
## ticks :바범위, shrink:바크기, orientation : 하단배치,옆배치
plt.colorbar(ticks=[1,2,3],label='학년',shrink=0.5\
  , orientation='horizontal')

plt.show()
```
