```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# 한글설정
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Win용
# matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac용
matplotlib.rcParams['font.size'] = 15 # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False

df = pd.read_excel('score.xlsx',index_col='지원번호')
```

```python
### 학교별 인원을 원그래프로 출력하시오.
count_df = df.groupby('학교')['이름'].count()
count_df.index
wedgeprops = {'width':0.6,'edgecolor':'w','linewidth':4}
# explode = [0.1,0]
plt.pie(count_df,labels=count_df.index,autopct='%.1f%%',startangle=90,
        wedgeprops=wedgeprops,pctdistance=0.7)
plt.show()
```
