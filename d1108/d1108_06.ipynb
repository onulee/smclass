```python
import pandas as pd
df = pd.read_excel('score.xlsx',index_col='지원번호')


# 학교 구로고->가산고, 디지털고->단지고 로 변경하시오.
df['학교'].replace({'구로고':'가산고','디지털고':'단지고'})
# SW특기 소문자로 변경하시오.
df['SW특기'] = df['SW특기'].str.lower()
# 4번 SW특기 -> django
# 5번 SW특기 -> python
df.loc['4번','SW특기'] = 'django'
df.loc[['4번','5번'],'SW특기'] = ['django','python']
# 합계컬럼을 추가하시오.
df['합계'] = df['국어'] + df['영어'] + df['수학'] + df['과학'] + df['사회']
# 결과컬럼을 추가해서 400점이상은 pass, 아니면 fail 출력하시오.
df['결과'] = 'Fail'
df.loc[df['합계'] > 400,'결과'] = 'Pass'
```

```python
### 문자열 = 문자열 + 문자열
df['학교'] = df['학교']+'등학교'
```
